```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df1=pd.read_csv("2_TwoNum.csv")
df1.head()
```

removing some data point for brtter visualisation

```python
df1=df1[df1["GrLivArea"]<=4000]
df1=df1[df1["SalePrice"]<=500000]
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,5))
plt.xticks(rotation=-45, ha='left')
sns.histplot(data=df1,x='GrLivArea',ax=axs[0])
sns.histplot(data=df1,x='SalePrice',ax=axs[1])
```

```python
sns.kdeplot(x=df1["GrLivArea"],y=df1["SalePrice"])
```

we can see high correlation in the data

```python
sns.regplot(x=df1["GrLivArea"],y=df1["SalePrice"],scatter_kws={"alpha":1,"s":2})
# alpha adjust the transparency of the points ,s denotes the size of the point 
```

```python
for i in range(5,11):
    sns.kdeplot(x=df1["GrLivArea"],y=df1["SalePrice"],cmap="Blues", shade=True, bw_adjust=i/10)
    plt.show()
```

```python
sns.jointplot("GrLivArea", "SalePrice", df1, kind='kde')
```
