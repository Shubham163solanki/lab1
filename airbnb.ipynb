```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df1=pd.read_csv("1_OneNum.csv")

df1.head()
```

removing all the airbnb with price greater than 300.

```python
df1=df1[df1['price']<=300]
```

```python
for i in range(10,50,5) :
    sns.histplot(data=df1,x='price',binrange=(0,300),binwidth=i)
    plt.show()
```

```python
plt.xlim(0,300)
sns.distplot(df1["price"],kde=False)
```

```python
sns.kdeplot(data=df1)
```
